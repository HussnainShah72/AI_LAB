```python

routes = {
    "Route A": {"distance": 10, "time": 15, "risk": "Low"},
    "Route B": {"distance": 8,  "time": 20, "risk": "Medium"},
    "Route C": {"distance": 12, "time": 10, "risk": "High"}
}

risk_values = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

def cal_utility(distance, time, risk):
    utility = (distance + time + risk)
    return utility

utilities = {}
best_route = None
max_utility = float("inf")

for route, attributes in routes.items():
    distance = attributes["distance"]
    time = attributes["time"]
    risk = risk_values[attributes["risk"]]

    utility = cal_utility(distance, time, risk)
    utilities[route] = utility

best_route = min(utilities,key=utilities.get)
        
print("Utility Scores:")
for route, utility in utilities.items():
    print(route, ":", utility)

print("\nSelected Route:", best_route)
```
